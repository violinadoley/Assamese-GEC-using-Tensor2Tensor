```python
import pandas as pd

file_path = "/kaggle/working/final_dataset.csv"  
df = pd.read_csv(file_path)

correct_sentences = df['Sentence']
incorrect_sentences = df['Incorrect Sentence']

with open("/kaggle/working/target.txt", "w", encoding="utf-8") as target_file:
    target_file.write("\n".join(correct_sentences))

with open("/kaggle/working/source.txt", "w", encoding="utf-8") as source_file:
    source_file.write("\n".join(incorrect_sentences))

print("Files target.txt and source.txt have been created successfully.")
```
